# Iterations in python 
1. iterations are loops to use perform the repeatative tasks or repeatative performances.
2. generally iterations are 2 types
3. `for` loop
4. `while` loop

```python - for loop
for iter_variables in sequence:
    print("for is keyword to start the iterations")
    print("iter_variables is variable to iterate data from sequence")
    print("in is keyword to validate the sequence is present or not")
    print("sequence is collection like range,string,list,tuple....etc")
```

```python - while loop
while (condition):
    print("while is keyword to start the loop and validation conditions based on values")
    print("If condition is true the inside statements are executed")
```

````python
#write a program to 1 to 60 numbers using loop?
for num in range(1,61):
    print(num)
````

````python
# write a program to iterate characters from string using for loop?
name="vikas kumar"
for char in name:
    print(char)
````

````python
#write  a program to print even number is between 500 to 600 using for loop?
for n in range(500,601):
    if (n%2 ==0):
        print(n,end=",")
````

````python
#write  a program to print odd number is between 500 to 600 using for loop?
for n in range(500,601):
    if (n%2 ==1):
        print(n,end=",")
````
